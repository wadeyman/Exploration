```python
from geopy.distance import geodesic

# def find_nearest(house, amenity):
#     """
#     this function finds the nearest locations from the 2nd table from the 1st address
#     Both are dataframes with a specific format:
#         1st column: any string column ie addresses taken from the "find_postal_address.py"
#         2nd column: latitude (float)
#         3rd column: longitude (float)
#     Column name doesn't matter.
#     """
#     results = {}
#     # first column must be address
#     for index,flat in enumerate(house.iloc[:,0]):
        
#         # 2nd column must be latitude, 3rd column must be longitude
#         flat_loc = (house.iloc[index,1],house.iloc[index,2])
#         flat_amenity = ['','',100]
#         for ind, eachloc in enumerate(amenity.iloc[:,0]):
#             amenity_loc = (amenity.iloc[ind,1],amenity.iloc[ind,2])
#             distance = geodesic(flat_loc,amenity_loc)

#             if distance < flat_amenity[2]:
#                 flat_amenity[0] = flat
#                 flat_amenity[1] = eachloc
#                 flat_amenity[2] = distance

#         results[flat] = flat_amenity
#     return results

def find_nearest(house, amenity):
    """
    This function finds the nearest locations from the 2nd table from the 1st address.
    Both are dataframes with a specific format:
        1st column: any string column i.e., addresses taken from the "find_postal_address.py"
        2nd column: latitude (float)
        3rd column: longitude (float)
    Column name doesn't matter.
    """
    results = {}

    # Convert DataFrames to NumPy arrays for faster computation
    house_array = house.values
    amenity_array = amenity.values

    def find_nearest_for_flat(flat_index):
        flat = house_array[flat_index, 0]
        flat_loc = (house_array[flat_index, 1], house_array[flat_index, 2])
        flat_amenity = ['', '', 100]

        for amenity_index in range(len(amenity_array)):
            eachloc = amenity_array[amenity_index, 0]
            amenity_loc = (amenity_array[amenity_index, 1], amenity_array[amenity_index, 2])
            distance = geodesic(flat_loc, amenity_loc).kilometers

            if distance < flat_amenity[2]:
                flat_amenity[0] = flat
                flat_amenity[1] = eachloc
                flat_amenity[2] = distance

        return flat, flat_amenity

    with ThreadPoolExecutor() as executor:
        # Map each flat index to its nearest amenity
        futures = [executor.submit(find_nearest_for_flat, i) for i in range(len(house_array))]

        for future in futures:
            flat, flat_amenity = future.result()
            results[flat] = flat_amenity

    return results
```
